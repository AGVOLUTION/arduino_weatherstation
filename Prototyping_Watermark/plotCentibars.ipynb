```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cbars(Rsense):
    TempC = 16
    if(Rsense <= 550):
        cbars = 0
    elif((Rsense > 550) and (Rsense <= 1000)):
        cbars = ( 23.156 * ( Rsense / 1000.0 ) - 12.736 ) * ( 1.0+ 0.018 * ( TempC - 24.0 ) )
    elif((Rsense > 1000) and (Rsense <= 8000)):
        cbars = ( -3.213 * ( Rsense / 1000.0 ) - 4.093 ) / ( 1.0- 0.009733 * ( Rsense / 1000.0 ) - 0.01205 * TempC)
    else:
        cbars = -2.246 - 5.239 * ( Rsense / 1000.0 ) * ( 1.0+ 0.018 * ( TempC - 24.0 ) ) - 0.06756 * ( Rsense / 1000.0 ) * ( Rsense / 1000.0 ) * ( ( 1.0 + 0.018 * (TempC-24.0)) * (1.0 + 0.018 * (TempC - 24.0)))
    return abs(round(cbars))
```

```python
rs = np.linspace(0, 39300, 500)
fix, ax = plt.subplots()
cbarsArr = np.zeros(len(rs))
for i in range(len(rs)):
    cbarsArr[i] = cbars(rs[i])
ax.plot(rs, cbarsArr)
plt.show()
print(max(cbarsArr))
```
